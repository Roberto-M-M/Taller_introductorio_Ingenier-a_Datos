8. Encuentre en su dataset una característica categórica (no numérica) con pocos valores u opciones diferentes y conviertala en un dato numérico con la función cat.codes. Recuerde que el resultado se almacene en una nueva columna.

```python
#Primero debemos cambiar el tipo de dato a un tipo categórico para que la función logre entender la transformación
datos_csv_url_clean['artist(s)_name'] = datos_csv_url_clean['artist(s)_name'].astype('category')
datos_csv_url_clean.dtypes
```

```python
datos_csv_url_clean['artist(s)_name'].value_counts()
```

```python
#Lo siguiente es agregar una nueva columna con los datos categorizados gracias a la función cat.codes de pandas
datos_csv_url_clean['Club_Art'] = datos_csv_url_clean['artist(s)_name'].cat.codes
datos_csv_url_clean.tail()
```

```python
#Comprobamos que se haya hecho la cotegorización:
datos_csv_url_clean[datos_csv_url_clean['Club_Art'] == 1]
```

10. Utilice los operadores relacionales para transformar una característica numérica en un dato categórico.

```python
# Definir la función de clasificación
def clasificacion(x):
    if x['streams'] > 320336326:
        return 'Escuchado'
    elif 32033632 < x['streams'] < 320336326:
        return 'Medio'
    else:
        return 'Sordo'

# Aplicar la función a cada fila del DataFrame
datos_csv_url_clean['Escuchados_world'] = datos_csv_url_clean.apply(clasificacion, axis=1)

# Mostrar los primeros registros del DataFrame con la nueva columna
datos_csv_url_clean.head()
```

```python
#Para escribir datos o guardar nuestro dataframe transformado basta con usar: 
# df.to_csv(r'Path where you want to store the exported CSV file\File Name.csv')
datos_csv_url_clean.to_csv('Datos_procesados_Spotify.csv', encoding='utf-8-sig') 
```

```python
import matplotlib.pyplot as plt

# Contar la frecuencia de cada categoría en la columna 'Escuchados_world'
frecuencia_categorias = datos_csv_url_clean['Escuchados_world'].value_counts()

# Crear la gráfica de barras
plt.bar(frecuencia_categorias.index, frecuencia_categorias.values)

# Agregar etiquetas y título
plt.xlabel('Categoría de audición')
plt.ylabel('Cantidad')
plt.title('Frecuencia de Categorías en Escuchados_world')

# Mostrar la gráfica
plt.show()
```
